```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pylab as plt
plt.style.use('ggplot')
```

```python
df_demo = sns.load_dataset('iris')
```

```python
df_demo
```

# Correlation

```python
# vars to analize
vars = ['sepal_length','sepal_width','petal_length','petal_width']
```

```python
# calculate correlation
corr = df_demo[vars].corr()
```

```python
# visualize correlation

fig, axs = plt.subplots(1, figsize=(10, 10), dpi=150, sharey=False)

sns.heatmap(
    corr,
    cmap= sns.color_palette("vlag", as_cmap=True),
    annot= True, # show each corr value
    ax= axs 
    )

axs.set_title('Correlation')
plt.show()
```
